```python

columnas_num = ['Área Cosechada (ha)','PRODUCCIÓN (Ton)']

fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(8,30))
fig.subplots_adjust(hspace=0.5)

for i, col in enumerate(columnas_num):
    sns.boxplot(data=datosA, x=col, ax=ax[i])
    ax[i].set_title(col)
```

### Errores Tipograficos


```python

cols_cat = ['Año', 'Departamento', 'Municipio', 'CULTIVO', 'Periodo', 'Ciclo']


fig, ax = plt.subplots(nrows=6, ncols=1, figsize=(10,30))
fig.subplots_adjust(hspace=5)

for i, col in enumerate(cols_cat):
  sns.countplot(x=col, data=datosA, ax=ax[i])
  ax[i].set_title(col)
  ax[i].set_xticklabels(ax[i].get_xticklabels(),rotation=90)
```

### 4.6 Porduccion por municipio

```python
datosA["PRODUCCIÓN (Ton)"] = (
    datosA["PRODUCCIÓN (Ton)"]
    .astype(str)
    .str.replace(",", "", regex=False)
    .str.replace(" ", "")
    .astype(float)
)


prod = datosA.groupby("CULTIVO")["PRODUCCIÓN (Ton)"].sum().sort_values(ascending=False).head(15)

plt.figure(figsize=(14,6))
prod.plot(kind="bar")
plt.xticks(rotation=45, ha="right")
plt.ylabel("Toneladas")
plt.title("Top 15 cultivos por producción total")
plt.tight_layout()
plt.show()

```

```python
plt.figure(figsize=(8,6))
plt.scatter(datosA["Rendimiento  (t/ha)"], datosA["PRODUCCIÓN (Ton)"])
plt.xlabel("Rendimiento (t/ha)")
plt.ylabel("Producción (Ton)")
plt.title("Rendimiento vs producción")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.scatter(datosA["Área Sembrada (ha)"], datosA["PRODUCCIÓN (Ton)"])
plt.xlabel("Área Sembrada (ha)")
plt.ylabel("Producción (Ton)")
plt.title("Relación entre área sembrada y producción")
plt.tight_layout()
plt.show()
```

```python
cultivos = datosA.groupby("CULTIVO").agg({
    "Área Sembrada (ha)": "sum",
    "PRODUCCIÓN (Ton)": "sum",
    "Rendimiento  (t/ha)": "mean"
}).sort_values(by="PRODUCCIÓN (Ton)", ascending=False)
print(cultivos)
```

```python
plt.figure(figsize=(12,5))
plt.bar(cultivos.index, cultivos["PRODUCCIÓN (Ton)"])
plt.xticks(rotation=45, ha="right")
plt.title("Producción total por cultivo")
plt.ylabel("Toneladas")
plt.show()


```

# 10 cultivos con mayor Rendimiento

```python
top = cultivos.sort_values(by="Rendimiento  (t/ha)", ascending=False).head(10)

plt.figure(figsize=(10,4))
plt.bar(top.index, top["Rendimiento  (t/ha)"])
plt.xticks(rotation=45, ha="right")
plt.title("Top 10 cultivos con mayor rendimiento")
plt.ylabel("Ton/Ha")
plt.tight_layout()
plt.show()
```
